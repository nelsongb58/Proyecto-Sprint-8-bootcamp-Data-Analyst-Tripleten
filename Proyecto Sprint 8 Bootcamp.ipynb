```python
import pandas as pd
df_trips = pd.read_csv('/datasets/project_sql_result_07.csv')
```

```python
print(df_trips.head(10))
print(df_trips.info())
```

```python
# Convertir la columna a formato datetime
df_trips['start_ts'] = pd.to_datetime(df_trips['start_ts']) 
# Extraer el día de la semana
df_trips['day_of_week'] = df_trips['start_ts'].dt.day_name()  
# Extraer el día de la semana
saturdays = df_trips[df_trips['day_of_week'] == 'Saturday']  
# Filtrar por condicion climatica
rainy_saturdays = saturdays[saturdays['weather_conditions'] == 'Bad'] 
```

Lo primero que se realizó fue agrupar los sábados en días lluviosos y no lluviosos, lo cual es un paso indispensable para llevar a cabo la prueba de hipótesis. Una vez completado este agrupamiento, se aplicó una prueba de hipótesis utilizando el método de t-Student, estableciendo un nivel de significancia de 0.05. A continuación, se presenta el código correspondiente junto con la interpretación de los resultados.

```python
from scipy.stats import ttest_ind, levene

# Prueba de Levene para comparar las varianzas
stat_levene, p_value_levene = levene(rainy_saturdays['duration_seconds'], saturdays['duration_seconds'])

# Nivel de significación
alpha = 0.05

# Verificar el resultado de la prueba de Levene
if p_value_levene > alpha:
    print("No rechazamos H₀ de Levene: Las varianzas son iguales.")
    equal_var = True
else:
    print("Rechazamos H₀ de Levene: Las varianzas no son iguales.")
    equal_var = False

# Realizar la prueba t considerando el resultado de Levene
stat_ttest, p_value_ttest = ttest_ind(rainy_saturdays['duration_seconds'], saturdays['duration_seconds'], equal_var=equal_var)

# Resultados de la prueba t
if p_value_ttest < alpha:
    print("Rechazamos H₀ del t-test: La duración promedio cambia significativamente en sábados lluviosos.")
else:
    print("No podemos rechazar H₀ del t-test: No hay evidencia suficiente para afirmar un cambio significativo.")

```
